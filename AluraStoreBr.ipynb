#1. Análise do faturamento


Nesta primeira análise, vamos calcular o faturamento total de cada loja. Somando os valores da coluna Preço de cada loja para estimar o faturamento.

```python
# Calculando o faturamento usando a coluna preco e somando toda
faturamento1 = loja['preco'].sum()
faturamento2 = loja2['preco'].sum()
faturamento3 = loja3['preco'].sum()
faturamento4 = loja4['preco'].sum()

print(f"Faturamento médio da Loja 1 é de R$ {faturamento1:,.2f}")
print(f"Faturamento médio da Loja 2 é de R$ {faturamento2:,.2f}")
print(f"Faturamento médio da Loja 3 é de R$ {faturamento3:,.2f}")
print(f"Faturamento médio da Loja 4 é de R$ {faturamento4:,.2f}")
```

```python
# Como calculamos anteriormente anteriormente, vamos verificar qual loja teve o maior faturamento
faturamentos = {
    'Loja 1': faturamento1,
    'Loja 2': faturamento2,
    'Loja 3': faturamento3,
    'Loja 4': faturamento4
}

# Calculando a loja com maior faturamento
maior_loja = max(faturamentos, key=faturamentos.get)
maior_faturamento = faturamentos[maior_loja]

menor_loja = min(faturamentos, key=faturamentos.get)
menor_faturamento = faturamentos[menor_loja]

# Print do resultado
print(f"O maior faturamento foi de R$ {maior_faturamento:,.2f}, correspondente à {maior_loja}.")
print(f"O menor faturamento foi de R$ {menor_faturamento:,.2f}, correspondente à {menor_loja}.")

```

# 2. Vendas por Categoria


Neste passo, vamos calcular a quantidade de produtos vendidos por categoria em cada loja. A ideia é agrupar os dados por categoria e contar o número de vendas de cada tipo, mostrando as categoria mais populares de cada loja

```python
# Agrupa e conta quantos produtos foram vendidos por categoria
qtd_vendidos_loja = loja.groupby('categoria_produto')['categoria_produto'].count()
qtd_vendidos_loja2 = loja2.groupby('categoria_produto')['categoria_produto'].count()
qtd_vendidos_loja3 = loja3.groupby('categoria_produto')['categoria_produto'].count()
qtd_vendidos_loja4 = loja4.groupby('categoria_produto')['categoria_produto'].count()

# Ordena em ordem decrescente
qtd_vendidos_loja = qtd_vendidos_loja.sort_values(ascending=False)
qtd_vendidos_loja2 = qtd_vendidos_loja2.sort_values(ascending=False)
qtd_vendidos_loja3 = qtd_vendidos_loja3.sort_values(ascending=False)
qtd_vendidos_loja4 = qtd_vendidos_loja4.sort_values(ascending=False)

# Print dos resultados
print("Vendas por categoria da Loja 1")
print(qtd_vendidos_loja)
print("Vendas por categoria da Loja 2")
print(qtd_vendidos_loja2)
print("Vendas por categoria da Loja 3")
print(qtd_vendidos_loja3)
print("Vendas por categoria da Loja 4")
print(qtd_vendidos_loja4)
```

# 4. Produtos Mais e Menos Vendidos

Neste passo, vamos identificar quais produtos foram os mais vendidos e os menos vendidos em cada loja.

```python
# Dicionário com as lojas
lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}
# Loop for para iterar no nome da loja e calcular o produto mais vendido e sua quantidade
for nome_loja, df_loja in lojas.items():
    contagem = df_loja['produto'].value_counts()

    mais_vendido = contagem.idxmax()
    mais_vendido_qtd = contagem.max()

    menos_vendido = contagem.idxmin()
    menos_vendido_qtd = contagem.min()

    print(f"[{nome_loja}] Produto mais vendido: {mais_vendido} ({mais_vendido_qtd} vendas)")
    print(f"[{nome_loja}] Produto menos vendido: {menos_vendido} ({menos_vendido_qtd} vendas)\n")
```

#6. Análise Geográfica



```python
# Mapeando as regiões do BR com base na sua UF
regioes = {
    'Norte': ['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO'],
    'Nordeste': ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE'],
    'Centro-Oeste': ['DF', 'GO', 'MT', 'MS'],
    'Sudeste': ['ES', 'MG', 'RJ', 'SP'],
    'Sul': ['PR', 'RS', 'SC']
}
```

```python
# Criando a coluna região para cada um dos DFs usando uma função
def mapear_regiao(uf):
    if uf in ['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO']:
        return 'Norte'
    elif uf in ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE']:
        return 'Nordeste'
    elif uf in ['DF', 'GO', 'MT', 'MS']:
        return 'Centro-Oeste'
    elif uf in ['ES', 'MG', 'RJ', 'SP']:
        return 'Sudeste'
    elif uf in ['PR', 'RS', 'SC']:
        return 'Sul'
    else:
        return 'Desconhecido'

loja['regiao'] = loja['local_compra'].apply(mapear_regiao)
loja2['regiao'] = loja2['local_compra'].apply(mapear_regiao)
loja3['regiao'] = loja3['local_compra'].apply(mapear_regiao)
loja4['regiao'] = loja4['local_compra'].apply(mapear_regiao)
```

```python
# Comparando o desempenho da loja1 por região:
print("Loja 1:")
loja.groupby('regiao').agg({
    'preco': 'mean',
    'frete': 'mean',
    'avaliacao_compra': 'mean'
}).rename(columns={
    'preco': 'Faturamento Médio',
    'frete': 'Frete Médio',
    'avaliacao_compra': 'Avaliação Média'
}).round(2)

```

```python
# Comparando o desempenho da loja3 por região:
print("Loja 3:")
loja3.groupby('regiao').agg({
    'preco': 'mean',
    'frete': 'mean',
    'avaliacao_compra': 'mean'
}).rename(columns={
    'preco': 'Faturamento Médio',
    'frete': 'Frete Médio',
    'avaliacao_compra': 'Avaliação Média'
}).round(2)
```

# 7. Gerando gráficos para visualização


Os gráficos vieram para complementar a análise realizada, destacando os pontos mais relevantes, como o faturamento das lojas, a distribuição das categorias de produtos, as avaliações dos clientes, os produtos mais vendidos, e ou custo do frete.

```python
lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}

# Calcula as categorias mais vendidas para todas as lojas
categorias_top = set()
for df_loja in lojas.values():
    categorias_top.update(df_loja['categoria_produto'].value_counts().index)

categorias_top = list(categorias_top)  # Converte para lista

# Dados organizados para o gráfico
dados = {categoria: [] for categoria in categorias_top}
for categoria in categorias_top:
    for df_loja in lojas.values():
        dados[categoria].append(df_loja['categoria_produto'].value_counts().get(categoria, 0))

# Gráfico de barras agrupadas
x = np.arange(len(categorias_top))  # Posições no eixo X
largura = 0.2  # Largura de cada barra

plt.figure(figsize=(12, 6))
for i, (nome_loja, valores) in enumerate(lojas.items()):
    plt.bar(x + i * largura, [dados[categoria][i] for categoria in categorias_top],
            width=largura, label=nome_loja)

plt.title('Categorias Mais Vendidas (Todas as Lojas)')
plt.ylabel('Quantidade Vendida')
plt.xlabel('Categorias de Produtos')
plt.xticks(x + largura * (len(lojas) - 1) / 2, categorias_top, rotation=45)
plt.legend(title='Lojas')
plt.tight_layout()
plt.show()
```

1. A Loja 1 tem o frete mais caro (R$34,6) e a pior avaliação média (3,98).

2. A Loja 3 tem a melhor avaliação (4,05) com um frete médio menor (R$33,0).

3. A Loja 4, com o frete mais barato (R$31,4), tem uma avaliação razoável (3,99), mas não a melhor.

Podemos interpretar qua há uma leve tendencia negativa onde quanto maior o frete, pior a avaliação. Sugerindo que fretes altos podem impactar negativamente na experiência do consumidor.

## **Com base no heatmap de correlações:**
As correlações são baixas (exceto a do frete com faturamento), então não há relações fortes, mas os pequenos sinais negativos reforçam o ponto anterior: frete alto pode impactar avaliação.

## **Com base no heatmap da distribuição dos pedidos:**
A maior concentração de pedidos da **loja 1** está provavelmente entre São Paulo, Rio de Janeiro e Curitiba. Regiões como Norte e parte do Centro-Oeste apresentam baixa densidade ou ausência de pedidos

Lojas 2 e 3 possuem distribuições semelhantes: concentrada no sudeste

Loja 4 também concentrada no Sudeste, mas há alguns pedidos mais ao Sul

## **Com base nas tabelas de regionais onde comparamos as lojas:**
1. **Olhando para faturamento:**

  Centro-Oeste lidera nas Lojas 1 e 3 com os maiores faturamentos médios:
    R$746,37 e R$754,01, respectivamente.

  Loja 2 tem o menor faturamento médio no Centro-Oeste (R$535,81), o que destoa fortemente das demais.

  Loja 4 tem o pior desempenho geral em faturamento médio, exceto no Norte e Sul.

2. **Olhando para o frete:**
  
  Em geral, a Loja 4 tem os menores valores de frete, com destaque para o Centro-Oeste (R$25,89). Isso pode ser devido a alguma ao baixo faturamento

  A Loja 1 e Loja 3 têm os fretes mais altos, principalmente no Centro-Oeste e Norte (acima de R$37).
  Esse fato pode estar ligado por conta das lojas, como vimos antes, estar no sudeste.

3. **Olhando para avaliação dos clientes:**

  O Sudeste e o Sul apresentam as melhores avaliações médias na maioria das lojas. Muito por conta da maioria dos pedidos vierem dessas regiões (o que pode enviesar).

  A Loja 2 e a Loja 4 têm os maiores valores de avaliação (acima de 4.0) em quase todas as regiões — mesmo com menores faturamentos — sugerindo que podem estar oferecendo melhor experiência ao cliente.

  A Loja 1 tem desempenho misto: maior faturamento em várias regiões, mas avaliações medianas (exceto no Sudeste).

  

## **Com base na análise por safra:**

Não há grandes diferenças entre as safras de todas as lojas, parecem uniformes, não evidenciando nenhuma tendência temporal.

## **Resumo**
1. Faturamento Médio
- Melhor desempenho: Loja 1
- Pior desempenho: Loja 4

A Loja 4 tem o menor faturamento médio em geral, ficando atrás das demais em quase todas as regiões, com exceção do Norte e Sul.

2. Vendas por Categoria
- Todas as lojas têm 'móveis' como categoria mais vendida, o que mostra consistência na preferência do público.
- A Loja 3 é a que mais vende móveis (499).
- A Loja 2 é a que vende menos móveis (442).

3. Avaliação Média Geral
- Melhor avaliação: Loja 3 (4.05)
- Segunda melhor: Loja 2 (4.04)
- Pior avaliação: Loja 1 (3.88)

Loja 4 possui uma avaliação média próxima da Loja 2 e 3, com bom desempenho nesse aspecto.

4. Frete Médio
- Mais caro: Loja 1 (R$ 34.69)

- Mais barato: Loja 4 (R$ 31.28)

A Loja 4 tem o menor custo de frete, o que ajuda nas avaliações, mas não está conseguindo converter isso em maiores vendas ou faturamento.


## **Conclusão:**
Apesar de ter boas avaliações e o menor frete médio, a Loja 4 apresenta:

1. Baixo faturamento, o que impacta diretamente na geração de receita;

2. Volume de vendas inferior à Loja 3;

3. Avaliação boa, mas não suficiente para compensar a performance fraca em faturamento;

4. Menor presença em regiões de alta densidade de pedidos, conforme os mapas de calor.

**Portanto, Sr João recomendo vender a loja 4.**
